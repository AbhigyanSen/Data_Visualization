```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Dataset = pd.read_csv("Downloads/marketing_campaign.csv", index_col = 0)
```

```python
Dataset.head()
```

```python
plt.figure(figsize = (23, 25))
Pt = sns.heatmap(Dataset.corr(), annot = True, cmap = 'tab20b', vmin = -1, vmax =1)
Pt.set_yticklabels(Pt.get_yticklabels(), rotation = 0, fontsize = 15)
Pt.set_xticklabels(Pt.get_xticklabels(), rotation = 90, fontsize = 15)
plt.show()
```
